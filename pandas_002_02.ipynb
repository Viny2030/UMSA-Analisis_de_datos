## **<font color="DarkBlue">Repositorios</font>**

<p align="justify">
Es posible ingerir datos desde un repositorio web como el <a href = "https://archive.ics.uci.edu/">Repositorio de Aprendizaje Autom치tico de UCI</a>, que contiene una amplia colecci칩n de conjuntos de datos utilizados para investigaciones y educaci칩n en Machine Learning. El siguiente c칩digo lo encontrar치 en este <a href="https://archive.ics.uci.edu/dataset/2/adult">link</a>.

```python
!pip install ucimlrepo
from ucimlrepo import fetch_ucirepo

# Obt칠n el dataset con id=2 (Adult dataset)
adult = fetch_ucirepo(id=2)

# Los datos (caracter칤sticas) ya est치n en un DataFrame de pandas
X = adult.data.features
y = adult.data.targets

# Combina las variables explicativas y la objetivo en un solo DataFrame
df = pd.concat([X, y], axis=1)
df.head()
```

<p align="justify">
Tambi칠n es posbile cargar el conjunto de datos usando <code>fetch_openml</code> de <code>scikit-learn</code>.
<br><br>
<a href="https://scikit-learn.org/stable/modules/generated/sklearn.datasets.fetch_openml.html">Documentaci칩n</a>

```python
from sklearn.datasets import fetch_openml

# Cargar el conjunto de datos Adult Census desde sklearn
adult_census = fetch_openml(name="adult", version=2, as_frame=True)

# El conjunto de datos se carga como un Bunch, de donde podemos extraer el DataFrame
df = adult_census.frame
df.head()
```

```python
# Cargar el conjunto de datos Adult Census desde sklearn versi칩n 3
adult_census = fetch_openml(name="adult", version=3, as_frame=True)

# El conjunto de datos se carga como un Bunch, de donde podemos extraer el DataFrame
df = adult_census.frame
df.head()
```
