 ## **<font color="DeepPink">Mapeos de meses</font>**

```python
# Mapear los meses y aplicarlo al 铆ndice
meses = {"Enero":"1", "Febrero":"2", "Marzo":"3", "Abril":"4", "Mayo":"5", "Junio":"6", "Julio":"7", "Agosto":"8", "Septiembre":"9", "Octubre":"10", "Noviembre":"11", "Diciembre":"12"}
df.index = df.index.map(meses)
```

```python
df
```

 ## **<font color="DeepPink">Concatenaci贸n por ejercicios</font>**

```python
# Generar una lista de DataFrame
lista = []
```

```python
df.columns # los nombres de las columnas del dataframe
```

```python
list(enumerate(df.columns))
```

```python
for i, j in enumerate(df.columns):
  dataframe = pd.DataFrame(df[j])
  dataframe = dataframe.rename(columns={j: "iibb"})
  dataframe["a帽o"] = 2000 + i
  dataframe["a帽o"] = dataframe["a帽o"].astype("str")
  dataframe = dataframe[["a帽o", "iibb"]]
  lista.append(dataframe)
```

```python
# Concatenar los elementos de la lista
iibb = pd.concat(lista)
```

```python
iibb
```

 ## **<font color="DeepPink">Impuesto Automotor</font>**

```python
data_automotor = "https://atm.mendoza.gov.ar/portalatm/zoneBottom/datosInteres/recaudacion/recaudacion_impuesto_automotor.jsp"
automotor = pd.read_html(data_automotor, encoding='latin1')
df_auto = automotor[0]
df_auto.columns = df_auto.loc[1]
df_auto.columns.name = None
df_auto = df_auto.loc[2:13]
df_auto.dropna(axis=1, how="all", inplace=True)
df_auto.set_index("Meses", inplace=True)
df_auto = df_auto.astype("float")
df_auto.index = df_auto.index.map(meses)
df_auto
```

```python
px.line(df_auto, x=df_auto.index, y=["2024","2023","2022","2021"])
```

 ## **<font color="DeepPink">Impuesto Inmobiliario</font>**

```python
px.line(df_inmo, x=df_inmo.index, y=["2024","2023","2022","2021"])
```
