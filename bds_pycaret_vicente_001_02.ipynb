 ## **<font color="DeepPink">Metricas</font>**

 ### **<font color="DeepPink">Accuracy</font>**

<p align="justify">
<b>Accuracy</b> es una mtrica comnmente utilizada para evaluar el rendimiento de un modelo de clasificacin. Esta mtrica mide la proporcin de instancias clasificadas correctamente por el modelo sobre el total de instancias en el conjunto de datos de prueba.
<br><br>
Accuracy se calcula dividiendo el nmero de predicciones correctas (tanto verdaderos positivos como verdaderos negativos) entre el nmero total de instancias en el conjunto de prueba. La frmula es la siguiente:
<br><br>
Accuracy = $\frac{(VP + VN)}{Total}$
<br><br>
Los trminos de la frmula:<br><br>
<ul align="justify">
<li><b>Verdaderos positivos (VP)</b>: Son los casos positivos que el modelo clasifica correctamente como positivos.</li><br>
<li><b>Verdaderos negativos (VN)</b>: Son los casos negativos que el modelo clasifica correctamente como negativos.</li><br>
</ul>
<p align="justify">
En otras palabras, accuracy nos dice cuntas de las predicciones del modelo son correctas en comparacin con todas las predicciones realizadas.
<br><br>
<mark>Accuracy es una mtrica til cuando las clases estn bien equilibradas</mark> en el conjunto de datos, es decir,<mark>cuando el nmero de instancias de una clase es aproximadamente igual al nmero de instancias de la otra clase</mark>. Sin embargo, accuracy puede ser engaosa si las clases estn desequilibradas, ya que un modelo puede alcanzar alta accuracy al simplemente predecir siempre la clase mayoritaria en un conjunto de datos desequilibrado, pero sin ser capaz de generalizar bien a otras clases.
<br><br>
En casos de conjuntos de datos desequilibrados, es importante considerar otras mtricas...

 **<font color="DeepPink">Es la medida ms directa de la calidad de los clasificadores. Es un valor entre 0 y 1. Cuanto ms alto, mejor. </font>**

 ### **<font color="DeepPink">Recall</font>**

<p align="justify">
<b>Recall</b> (tambin conocido como sensibilidad o tasa de verdaderos positivos) es una mtrica utilizada para evaluar el rendimiento de un modelo de clasificacin, especialmente cuando el objetivo es identificar correctamente casos positivos en un conjunto de datos.
<br><br>
Recall se define como la proporcin de casos positivos que el modelo ha identificado correctamente sobre el total de casos positivos en el conjunto de datos. En otras palabras, es la capacidad del modelo para "recordar" o "capturar" correctamente los casos de la clase positiva.
<br><br>
La frmula para calcular el Recall es la siguiente:
<br><br>
Recall = $\frac{VP}{VP + FN}$
<br><br>
Los trminos de la frmula:<br><br>
<ul align="justify">
<li><b>Verdaderos positivos (VP)</b>: Son los casos positivos que el modelo clasifica correctamente como positivos.</li><br>
<li><b>Falsos negativos (FN)</b>: Son los casos positivos que el modelo clasifica incorrectamente como negativos</li><br>
</ul>
<p align="justify">
Recall mide la capacidad del modelo para encontrar y capturar todos los casos positivos en el conjunto de datos. Es especialmente til en problemas donde la deteccin de casos positivos es crucial, como en aplicaciones mdicas donde la identificacin de enfermedades o en problemas de deteccin de fraudes.
<br><br>
Es importante tener en cuenta que Recall y Accuracy estn relacionados, pero tambin pueden ser contradictorios. A medida que el modelo se vuelve ms sensible y mejora su Recall al capturar ms casos positivos, puede aumentar la cantidad de falsos positivos (casos negativos que se clasifican incorrectamente como positivos), lo que podra disminuir la precisin.
<br><br>
Por lo tanto, la eleccin de la mtrica adecuada depende del contexto y los objetivos del problema. Si se prioriza la identificacin correcta de los casos positivos, Recall es una mtrica clave a considerar.

**Es el nmero de elementos identificados correctamente como positivos del total de positivos verdaderos.**

 ### **<font color="DeepPink">Precision</font>**

<p align="justify">
<b>Precision</b> (tambin conocido como valor predictivo positivo) es otra mtrica importante utilizada para evaluar el rendimiento de un modelo de clasificacin. La Precision se enfoca en la proporcin de instancias clasificadas como positivas por el modelo que realmente son verdaderos positivos, es decir, la precisin en las predicciones positivas del modelo.
<br><br>
La frmula para calcular la Precision es la siguiente:
<br><br>
Precision = $\frac{VP}{VP + FP}$
<br><br>
Los trminos de la frmula significan lo siguiente:<br><br>
<ul align="justify">
<li><b>Verdaderos positivos (VP)</b>: Son los casos positivos que el modelo clasifica correctamente como positivos.</li><br>
<li><b>Falsos positivos (FP)</b>: Son los casos negativos que el modelo clasifica incorrectamente como positivos</li><br>
</ul>
<p align="justify">
En otras palabras, la Precision mide la proporcin de las instancias clasificadas como positivas que realmente son casos positivos entre todas las instancias clasificadas como positivas.
<br><br>
La Precision es particularmente relevante en situaciones donde el costo de los falsos positivos es alto, es decir, cuando clasificar incorrectamente una instancia como positiva tiene consecuencias graves o costosas. Un alto valor de Precision indica que el modelo es preciso en sus predicciones positivas y minimiza los errores de clasificacin de casos negativos como positivos.
<br><br>
Es importante tener en cuenta que la Precision y el Recall (sensibilidad) estn relacionados y suelen ser compensatorios entre s. A medida que se mejora la Precision al reducir los falsos positivos, es probable que disminuya el Recall, lo que implica que se pueden perder algunos verdaderos positivos. Por lo tanto, encontrar el equilibrio adecuado entre la Precision y el Recall depende de las necesidades y objetivos especficos del problema.
<br><br>
En resumen, la Precision es una mtrica crucial en la evaluacin del rendimiento de un modelo de clasificacin, especialmente en escenarios donde la precisin en las predicciones positivas es de gran importancia.
<br><br>
 Sin embargo, para obtener una visin completa del rendimiento del modelo, es necesario considerar otras mtricas, como por ejemplo: el F1-score, que combina la Precision y el Recall en una nica medida.

**Es el nmero de elementos identificados correctamente como positivo de un total de elementos identificados como positivos:**

 ### **<font color="DeepPink">F1</font>**

<p align="justify">
<b>F1</b> (tambin conocido como F-score o medida F) es una mtrica utilizada en la evaluacin de modelos de clasificacin binaria que combina Precision y el Recall en una nica medida.
<br><br>
Esta mtrica es til cuando se desea encontrar un equilibrio entre la precisin de las predicciones positivas y la capacidad de identificar correctamente los casos positivos en el conjunto de datos.
<br><br>
El F1-score se calcula como la media armnica entre la Precision y el Recall y se define mediante la siguiente frmula:
<br><br>
F1 = $2 \times \frac{Precision \times Recall}{Precision + Recall}$
<br><br>
La media armnica se utiliza para combinar Precision y Recall en el F1-score porque penaliza ms fuertemente los valores bajos. Esto significa que si una de las dos mtricas tiene un valor bajo, el F1-score tambin ser bajo, lo que destaca situaciones donde el modelo tiene un desequilibrio significativo entre Precision y Recall.
<br><br>
<mark>F1-score proporciona una medida ms equilibrada del rendimiento del modelo en comparacin con la precisin o el recall por separado, especialmente cuando las clases estn desequilibradas</mark> en el conjunto de datos. En problemas donde la precisin de las predicciones positivas y la capacidad de detectar correctamente los casos positivos son igualmente importantes, el F1-score es una mtrica valiosa.
<br><br>
Es esencial considerar el contexto y los objetivos del problema al elegir la mtrica de evaluacin ms adecuada.
<br><br>
 Si se da prioridad a la identificacin correcta de los casos positivos, el Recall es relevante.<br>
 Si se necesita alta precisin en las predicciones positivas y la identificacin de casos positivos no es tan crucial, la Precision puede ser ms relevante.<br>
 Sin embargo, si se busca un equilibrio entre ambas, el F1-score es la mtrica.

Siempre es mejor establecer como cientfico de datos una mtrica de evaluacin de un solo nmero para que su equipo luego optimice.

La precisin es un ejemplo de una mtrica de evaluacin de un solo nmero y le permite comparar rpidamente dos clasificadores.

 ## **<font color="DeepPink">Curva ROC</font>**

<p align="justify">
La Curva ROC (Receiver Operating Characteristic) es una herramienta grfica que se utiliza para evaluar el rendimiento de un modelo de clasificacin binaria.
<br><br>
Es una representacin visual de la sensibilidad (tasa de verdaderos positivos) frente a la tasa de falsos positivos $(1 - especificidad)$ para diferentes umbrales de decisin.
<br><br>
Aqu est la explicacin paso a paso de cmo se construye la Curva ROC:
<br><br>
<ol align="justify">
<li><b>Definicin de trminos</b>:</li>
<ul align="justify">
<li><b>Verdaderos positivos (VP)</b>: Son los casos positivos que el modelo clasifica correctamente como positivos.</li><br>
<li><b>Falsos positivos (FP)</b>: Son los casos negativos que el modelo clasifica incorrectamente como positivos.</li><br>
<li><b>Verdaderos negativos (VN)</b>: Son los casos negativos que el modelo clasifica correctamente como negativos.</li><br>
<li><b>Falsos negativos (FN)</b>: Son los casos positivos que el modelo clasifica incorrectamente como negativos</li><br>
</ul>
<li><b>Variacin del umbral de decisin</b>:
   Un modelo de clasificacin binaria asigna probabilidades o puntajes a cada instancia para determinar si pertenece a la clase positiva o negativa. Al cambiar el umbral de decisin, es posible ajustar el equilibrio entre la sensibilidad y la especificidad del modelo.</li><br>
<li><b>Clculo de la sensibilidad y especificidad</b>:
   Para cada umbral de decisin, se calculan la sensibilidad y la especificidad del modelo. La sensibilidad se define como la tasa de verdaderos positivos $\frac {VP} {VP + FN}$, mientras que la especificidad se define como la tasa de verdaderos negativos $\frac {VN}{VN + FP}$.</li><br>
<li><b>Construccin de la Curva ROC</b>:
   La Curva ROC se crea trazando la sensibilidad en el eje $Y$ (eje vertical) y el complemento de la especificidad $(1 - especificidad)$ en el eje $X$ (eje horizontal). A medida que el umbral de decisin vara, se obtienen diferentes puntos en la curva.</li><br>
<li><b>Interpretacin de la Curva ROC</b>:
   La Curva ROC muestra cmo vara el equilibrio entre la sensibilidad y la especificidad a diferentes niveles de umbral de decisin. Un modelo con un rendimiento excelente tendra una curva ROC que se acerca al vrtice superior izquierdo del grfico, lo que indica una alta sensibilidad y una baja tasa de falsos positivos. Un modelo sin capacidad de discriminacin tendra una curva ROC que se asemeja a la lnea diagonal (tasa de verdaderos positivos igual a tasa de falsos positivos).</li></ol>
<br>
<p align="justify">
Adems de la Curva ROC, se utiliza el rea bajo la curva (AUC, Area Under the Curve) como una mtrica resumen del rendimiento del modelo. El AUC mide la capacidad de discriminacin del modelo en general, y un valor de AUC cercano a $1$ indica un mejor rendimiento del modelo. En cambio, un valor cercano a $0.5$ indica que el modelo no supera el rendimiento aleatorio y, por lo tanto, es poco til.

 ## **<font color="DeepPink">Importancia de Variables</font>**

<p align="justify">
La "Feature Importance" o importancia de variables o caracteristicas, es un concepto utilizado en el contexto del aprendizaje automtico (machine learning) y, en particular, en modelos de clasificacin o regresin. Se refiere a una tcnica que permite determinar la contribucin relativa de cada caracterstica o variable utilizada en el modelo para realizar las predicciones.
<br><br>
Cuando se entrena un modelo de aprendizaje automtico, como un algoritmo de clasificacin (por ejemplo, Random Forest, Gradient Boosting, etc.) o un algoritmo de regresin (por ejemplo, Regresin Lineal, Lasso, etc.), el modelo utiliza un conjunto de caractersticas o variables (tambin conocidas como atributos o columnas) para hacer predicciones.
<br><br>
La importancia de caractersticas nos ayuda a responder preguntas como:
<br><br>
<ul aling="justify">
<li>Cules son las caractersticas ms influyentes para el modelo en sus predicciones?</li><br>
<li>Cules caractersticas pueden ser descartadas sin afectar significativamente el rendimiento del modelo?</li><br>
<li>Cules caractersticas tienen una correlacin ms fuerte con la variable objetivo (o la variable a predecir)?</li></ul><br>
<p align="justify">
Hay varias formas de calcular la importancia de caractersticas, y algunos algoritmos ya proporcionan esta informacin directamente despus de ser entrenados.
<br><br>
Algunas de las tcnicas comunes para calcular la importancia de caractersticas son:
<br>
<ol align="justify">
<li><b>Importancia basada en rboles (Tree-based Importance)</b>:
   En algoritmos de rboles, como Random Forest o Gradient Boosting, la importancia de caractersticas se puede calcular al observar cmo cada caracterstica afecta la precisin de la prediccin en el rbol. Las caractersticas que se utilizan para tomar decisiones en los niveles superiores del rbol o <mark>que reducen la impureza</mark> (como la entropa o la ganancia de informacin) en mayor medida, generalmente se consideran ms importantes.</li><br>
<li><b>Coeficientes en regresin lineal</b>:
   En modelos de regresin lineal o regresin lineal regularizada (por ejemplo, Lasso, Ridge), los coeficientes asociados con cada caracterstica indican su contribucin a la prediccin. Los coeficientes ms grandes, en valor absoluto, indican que esa caracterstica tiene una mayor influencia en las predicciones.</li><br>
<li><b>Permutacin de caractersticas (Permutation Feature Importance)</b>:
   En este enfoque, se mide cmo cambia el rendimiento del modelo cuando se permutan aleatoriamente los valores de una caracterstica especfica mientras se evala en un conjunto de prueba. <mark>Si una caracterstica es importante, su permutacin causar una disminucin significativa en el rendimiento del modelo</mark>.</li></ol><br>
<p align="justify">
La importancia de caractersticas es una herramienta til para comprender y analizar el comportamiento de un modelo de aprendizaje automtico, as como para identificar qu caractersticas tienen un mayor impacto en las predicciones.
<br><br>
Sin embargo, debe utilizarse con precaucin, ya que la interpretacin de la importancia puede variar segn el contexto y el problema especfico que se est abordando.
