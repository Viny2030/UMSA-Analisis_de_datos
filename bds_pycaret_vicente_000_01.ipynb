<li><b> Hay que tener en cuenta que en Aprendizaje Automático hay dos conceptos que parecen similares, aunque son completamente diferentes: </b>:

●	Parametro

●	Hiperparametro

<li><b> Ambos conceptos son aplicables en conjuntos de datos que se tienen que mejorar, a la hora de crear un modelo para un problema dado. </b>



<li><b> que es un parametro?</b>: Los parámetros son variables que se estiman durante el proceso de entrenamiento de un conjunto de datos. Sus valores no los indica manualmente el científico de datos, sino que son obtenidos. Los parámetros son la parte más importante en los modelos de Aprendizaje Automático, y es la parte de los modelos que aprende de los datos. Son necesarios para realizar las predicciones. Además de definir la capacidad del modelo para resolver un problema dado. Estimarlos correctamente es una tarea clave



<li><b> que es un hiperparametro?</b>: Los hiperparámetros de un modelo son los valores de las configuraciones utilizadas durante el proceso de entrenamiento. Son valores que generalmente no se obtienen de los datos, por lo que suelen ser indicados por el científico de datos

 # **<font color="DeepPink">¿Qué es PyCaret?</font>**

<p align="justify">
<code>PyCaret</code> es una biblioteca de código abierto para Python diseñada para facilitar y acelerar el proceso de desarrollo de modelos en Ciencia de Datos y Aprendizaje Automático (Machine Learning). Está especialmente diseñada para estudiantes y profesionales de Ciencia de Datos que deseen reducir el tiempo dedicado a tareas repetitivas y tediosas, como la preparación de datos, la selección de características, la optimización de hiperparámetros y la comparación de modelos.
<br><br>
Las principales características y ventajas de <code>PyCaret</code> son las siguientes:
</p><br>
<ol align="justify">
<li><b>Automatización de tareas</b>: <code>PyCaret</code> automatiza gran parte del flujo de trabajo de Ciencia de Datos, lo que permite a los estudiantes enfocarse más en la interpretación de los resultados y la toma de decisiones informadas.</li>
<br>
<li><b>Fácil de usar</b>: La biblioteca proporciona una interfaz sencilla y coherente que permite a los estudiantes crear y comparar múltiples modelos de aprendizaje automático con solo unas pocas líneas de código.</li>
<br>
<li><b>Preprocesamiento de datos</b>: <code>PyCaret</code> maneja automáticamente tareas como el manejo de valores faltantes, codificación de variables categóricas, normalización de datos y partición en conjuntos de entrenamiento y prueba.</li>
<br>
<li><b>Comparación de modelos</b>: <code>PyCaret</code> ofrece una tabla de clasificación donde los modelos se comparan automáticamente en función de diversas métricas de rendimiento, lo que ayuda a seleccionar el modelo más adecuado para el problema en cuestión.</li>
<br>
<li><b>Optimización de hiperparámetros</b>: La biblioteca realiza una búsqueda inteligente en el espacio de hiperparámetros para encontrar la combinación óptima que maximice el rendimiento del modelo.</li>
<br>
<li><b>Análisis de resultados</b> <code>PyCaret</code> proporciona herramientas para analizar y visualizar los resultados de manera sencilla, lo que facilita la interpretación de los modelos y la identificación de áreas de mejora</li>
<br>
<li><b>Compatibilidad con múltiples bibliotecas</b>: <code>PyCaret</code> es compatible con varias bibliotecas populares de aprendizaje automático, como <code>scikit-learn</code>, <code>XGBoost</code>, <code>LightGBM</code>, <code>CatBoost</code>, entre otras.</li>
<br>
<li><b>Flexibilidad y personalización</b>: A pesar de su enfoque en la automatización, <code>PyCaret</code> también permite a los estudiantes personalizar y ajustar los modelos y procesos según sea necesario.</li>
</ol>
<br>
<p align="justify">
En resumen, <code>PyCaret</code> es una herramienta valiosa para apasionados de Ciencia de Datos, ya que les permite aprovechar el poder del aprendizaje automático de manera rápida y eficiente, sin sacrificar el entendimiento y la interpretación de los resultados. Además, es una excelente manera de familiarizarse con diferentes técnicas de aprendizaje automático y construir modelos sólidos para resolver problemas del mundo real.

# **<font color="DeepPink"> Detalles del Setup</font>**

**<font color="DeepPink"> Session id</font>** Propiedad se usa para identificar de forma única un explorador con datos de sesión en el servidor

 **<font color="DeepPink"> Target</font>**-variable objetivo algoritmo de clasificacion binaria-"deposit"


**<font color="DeepPink">Target mapping</font>** = Mapeo de objetivo-si-no


 **<font color="DeepPink">Original data shape </font>** = formato de datos original(45211 filas, 17 columnas)

 **<font color="DeepPink">Transformed data shape  </font>** = formato de datos original transformado = (45211 filas, 19 columnas)


**<font color="DeepPink">Transformed train data shape  </font>** formato de datos de entrenamiento transformados (31647 filas, 49 columnas) -70% de datos


**<font color="DeepPink">Transformed test data shape  </font>**formato de datos de entrenamiento transformados (31647 filas, 49 columnas)-30 % de datos


 **<font color="DeepPink">Numeric features  </font>** = variables numericas

 **<font color="DeepPink">Categorical features  </font>** = variables categoricas

 **<font color="DeepPink">Preprocess </font>** = Preprocesamiento True

 **<font color="DeepPink">Imputation type  </font>** = Tipo de imputacion de las variables numericas simple

 **<font color="DeepPink">Categorical imputation  </font>** = Imputacion Categorica

 **<font color="DeepPink">Fold Generator  </font>** = metodo de division de separacion del conjunto de entrenamiento y testeo

 **<font color="DeepPink">Fold Number  </font>** = numero de pliegues en validacion cruzada. La validación cruzada es una técnica para evaluar modelos de ML mediante el entrenamiento de varios modelos de ML en subconjuntos de los datos de entrada disponibles y evaluarlos con el subconjunto complementario de los datos.Numero de pliegues que va a generar entrenamiento y testeo

Algunas tareas de computación, como la renderización de gráficos, el machine learning (ML) y la edición de videos, requieren la aplicación de operaciones matemáticas similares en un conjunto de datos grande. El diseño de una GPU le permite realizar la misma operación en varios valores en paralelo. Esto vuelve más eficiente el procesamiento de muchas tareas que requieren mucha capacidad de cómputo.
