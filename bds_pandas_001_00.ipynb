 ## **<font color="#07a8ed">Selecci칩n por etiquetas</font>**

<p align="justify">
<code>Pandas</code> proporciona un conjunto de m칠todos para lograr una indexaci칩n. Este es un protocolo estricto basado en la inclusi칩n. En el caso de indexaci칩n por etiquetas este es un protocolo estricto basado en la inclusi칩n. Cada etiqueta solicitada debe existir, tanto en el indice como en las columnas. Al seleccionar por ejemplo filas (indice) se incluye el l칤mite inicial indicado y el l칤mite final. <br><br> 游 Los n칰meros enteros son etiquetas v치lidas, pero se refieren a las etiquetas del indice (cuando la etiqueta es num칠rica) y no a la posici칩n de cada fila. Lo mismo pasa cuando en el 칤ndice tenemos fechas.
</p>

~~~python
df.loc[fila inicial : fila final, columna inicial : columna final]
~~~

 ## **<font color="#07a8ed">Selecci칩n por posici칩n</font>**

<p align="justify">
<code>Pandas</code> proporciona un conjunto de m칠todos para obtener una indexaci칩n puramente basada en enteros. La sem치ntica sigue de cerca lo definido en <code>Python</code> y <code>NumPy</code>. Al seleccionar, se incluye el l칤mite inicial $n$, mientras que se excluye el l칤mite final $m$, es decir desde $n$ hasta $m-1$ <br><br> 游 En los par치metros de la selecci칩n, siempre se indica primero las filas y luego las columnas. En el caso de seleccionar de un DataFrame, se hace de la siguiente forma:
</p>

~~~python
df.iloc[fila inicial : fila final, columna inicial : columna final]
~~~


<p align="justify">

 ## **<font color="#07a8ed">Filtros</font>**

<p align="justify">
<code>Pandas</code> nos permite llevar a cabo algunas opciones para filtrar nuestros datos, incluidas las m치scaras booleanas y algunos m칠todos especiales. Con las m치scaras booleanas, probamos nuestros datos contra alg칰n valor y obtenemos una estructura de la misma forma, excepto que est치 llena de valores que son <code>True</code> o <code>False</code>.
</p>


````python
df
````

### **<font color="#07a8ed"> Mayor ```>``` Mayor igual ```>=``` Menor ```<``` Menor igual ```<=```</font>**

<p align="justify">
Filtramos los valores Radio mayores a $40$.</p>


````python
df[df.Radio > 40]
````

<p align="justify">
Ahora filtramos los valores Radio mayores a $40$, pero con el m칠todo <code>query()</code>.</p>


````python
df.query("Radio > 40")
````

### **<font color="#07a8ed">Dos o m치s condiciones verdaderas ```&```</font>**

<p align="justify">
Para visualizar m치s de dos condiciones verdaderas, es decir todas las condiciones verdaderas se usa el s칤mbolo &...</p>


````python
df
````

````python
df[(df.Periodico > 11) & (df.Radio > 41)]
````

````python
df.query("Periodico > 11 & Radio > 41")
````

````python
df.query("Periodico > 11 and Radio > 41")
````

### **<font color="#07a8ed">Dos o m치s condiciones, alguna verdadera ```|```</font>**

<p align="justify">
Para visualizar m치s de dos condiciones y alguna de ellas es verdadera, se usa el s칤mbolo |...</p>


````python
df
````

````python
df[(df.Periodico > 15) | (df.Radio > 41)]
````

````python
df.query("Periodico > 15 | Radio > 41")
````

````python
df.query("Periodico > 15 or Radio > 41")
````

### **<font color="#07a8ed">Conjunto de valores ```between()```</font>**

<p align="justify">
Para un conjunto de valores entre un intervalo determinado...</p>


````python
df
````

````python
df[df.Periodico.between(10.4, 22.1)]
````

````python
df.query("Periodico >= 10.4 & Periodico <= 22.1")
````

````python
df.query("Periodico >= 10.4 and Periodico <= 22.1")
````

````python
df[df.Periodico.between(10.4, 22.1, inclusive="neither")] # excluye los extremos
````

````python
df.query("Periodico > 10.4 & Periodico < 22.1") # excluye los extremos
````
