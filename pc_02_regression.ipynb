## **Conjuntos de datos**

```python
from pycaret.datasets import get_data
index = get_data('index')
```

##**Carga de un conjunto de datos determinado**

***Creacin del DataFrame para un conjunto de datos determinado, como por ejemplo el conjunto de datos "insurance"***

```python
data = get_data('insurance')
```

```python
from pycaret.regression import *
```

```python
setup = setup(data, target = 'charges')
```

```python
mejor_modelo = compare_models()
```

```python
gbr = create_model('gbr')
```

https://scikit-learn.org/stable/modules/generated/sklearn.ensemble.GradientBoostingRegressor.html


### Hyperparametros:
class sklearn.ensemble.GradientBoostingRegressor(*, loss='squared_error', learning_rate=0.1, n_estimators=100, subsample=1.0, criterion='friedman_mse', min_samples_split=2, min_samples_leaf=1, min_weight_fraction_leaf=0.0, max_depth=3, min_impurity_decrease=0.0, init=None, random_state=None, max_features=None, alpha=0.9, verbose=0, max_leaf_nodes=None, warm_start=False, validation_fraction=0.1, n_iter_no_change=None, tol=0.0001, ccp_alpha=0.0)


```python
tune_gbr = tune_model(gbr)
```

```python
evaluate_model(tune_gbr)
```

```python
plot_model(tune_gbr)
```

```python
plot_model(tune_gbr, plot = 'error')
```

```python
plot_model(tune_gbr, plot = 'feature')
```

```python
plot_model(tune_gbr, plot = 'learning')
```

```python
plot_model(tune_gbr, plot = 'residuals')
```

```python
 predict_model(tune_gbr)
```

```python
prediccion = predict_model(tune_gbr, data=data)
```

```python
prediccion
```

```python
prediccion['error_absoluto'] = abs(prediccion['charges'] - prediccion['prediction_label'])
```

```python
prediccion['error_absoluto'].mean()
```
