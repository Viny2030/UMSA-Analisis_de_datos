 # **<font color="DarkBlue">ETL versus ELT</font>**

<p align="justify">
La diferencia entre <b>ETL</b> (Extract, Transform, Load) y <b>ELT</b> (Extract, Load, Transform) radica principalmente en el orden en que se ejecutan las etapas del proceso y en el entorno donde se llevan a cabo las transformaciones de los datos:

 ## **<font color="DarkBlue">ETL (Extract, Transform, Load)</font>**


1. **Extract (Extracci√≥n):** Se extraen los datos desde diversas fuentes (bases de datos, archivos, APIs, etc.).
2. **Transform (Transformaci√≥n):** Los datos se transforman en un entorno intermedio antes de ser cargados en el destino final. Este paso puede incluir la limpieza de datos, la agregaci√≥n, el filtrado y la transformaci√≥n en un formato m√°s adecuado para el an√°lisis.
3. **Load (Carga):** Los datos transformados se cargan en el sistema de almacenamiento o base de datos destino, como un data warehouse.

üöÄ **Caracter√≠sticas clave de ETL:**
- La transformaci√≥n se realiza antes de la carga.
- Es com√∫n en sistemas tradicionales de almacenamiento de datos, como **data warehouses**.
- A menudo se utiliza cuando los datos que se cargan deben ser procesados previamente para cumplir con un esquema espec√≠fico o para optimizar el an√°lisis.



 ## **<font color="DarkBlue">ELT (Extract, Load, Transform)</font>**


1. **Extract (Extracci√≥n):** Los datos se extraen desde las fuentes.
2. **Load (Carga):** Los datos sin procesar se cargan directamente en el sistema de almacenamiento, que suele ser un **data lake** o una base de datos moderna.
3. **Transform (Transformaci√≥n):** Los datos se transforman dentro del propio sistema de almacenamiento o en el destino, utilizando la potencia de procesamiento del entorno, como un data warehouse o un motor de bases de datos.

üöÄ **Caracter√≠sticas clave de ELT:**
- La transformaci√≥n se realiza despu√©s de la carga.
- Es com√∫n en infraestructuras basadas en **data lakes** o en la nube, donde los sistemas pueden procesar grandes vol√∫menes de datos sin necesidad de transformar antes de cargar.
- A menudo se usa cuando el destino es lo suficientemente potente como para manejar grandes vol√∫menes de datos y realizar transformaciones a escala.



 ## **<font color="DarkBlue">Res√∫men de las diferencias claves:</font>**

- En **ETL**, los datos se transforman antes de ser cargados en el destino.
- En **ELT**, los datos se cargan primero y luego se transforman en el destino.

Por lo tanto:
- **ETL** es m√°s adecuado para sistemas que requieren datos bien estructurados antes del an√°lisis,
- mientras que **ELT** es ideal para entornos en la nube o sistemas que pueden manejar y procesar grandes vol√∫menes de datos sin estructurar.

<p align="justify"><br>
üëÄ Ambos enfoques son utilizados seg√∫n las necesidades de la organizaci√≥n y las caracter√≠sticas de los sistemas de almacenamiento.

<p align="justify">
‚úÖ El m√©todo <code>history</code> al igual de <code>dowload</code> devuelve un DataFrame de Pandas que contiene los datos hist√≥ricos del activo financiero especificado, con columnas como 'Open', 'High', 'Low', 'Close', 'Volume', etc. Algunos de sus hiperpar√°metros m√°s importantes son:

- `period` (str): El per√≠odo de tiempo para el que se desean los datos. Puede ser '1d', '1mo' o '1y'. Tambi√©n se puede especificar un per√≠odo personalizado utilizando el formato '2d', '5d', '1wk', '1mo', '3mo', '6mo', '1y', '2y', '5y', '10y', 'ytd' (Year to Date), o un rango espec√≠fico de fechas ('yyyy-mm-dd','yyyy-mm-dd'). Por defecto el per√≠odo de tiempo es 1 mes ('1mo').

- `interval` (str): El intervalo de tiempo para los datos. Puede ser '1m' para minutos, '5m' para 5 minutos, '15m' para 15 minutos, '30m' para 30 minutos, '60m' para 60 minutos, '90m' para 90 minutos, '1h' para 1 hora, '1d' para 1 d√≠a, '5d' para 5 d√≠as, '1wk' para 1 semana, o '1mo' para 1 mes. Por defecto el intervalo o periodicidad de los datos es diario.

- `start` (str o datetime): La fecha de inicio de los datos hist√≥ricos. Puede ser especificada como un string en formato 'yyyy-mm-dd' o como un objeto datetime de Python. Si no se proporciona, se utilizar√° la fecha m√°s temprana disponible.

- `end` (str o datetime): La fecha de fin de los datos hist√≥ricos. Puede ser especificada como un string en formato 'yyyy-mm-dd' o como un objeto datetime de Python. Si no se proporciona, se utilizar√° la fecha m√°s reciente disponible.

- `actions` (bool): Un booleano que indica si se deben incluir eventos como dividendos y divisiones de acciones en los datos hist√≥ricos. Por defecto es True.
