<p align="justify">
En este cuaderno comenzaremos a desarrollar los temás más usuales y utilizados del Análisis Exploratorio de los Datos, iniciando con la visualización de los datos.

## **<font color="#07a8ed">Bibliotecas**

### **<font color="#07a8ed">Para análisis de datos**

```
import pandas as pd
import numpy as np
```

### **<font color="#07a8ed">Para gráficos**

```
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

 ## **<font color="#07a8ed">Obtención de los Datos</font>**

```
url = "https://raw.githubusercontent.com/LucaAPiattelli/Diplomatura_Business_Analytics_UDA/main/Modulo_08_Aprendizaje_Automatico/visualizacion.csv"
```

```
analisis = pd.read_csv(url)
analisis.head()
```

Se cambian los nombres de las columnas para mejor interpretacion.

```
analisis.info()
```

```
analisis.rename(columns={"satisfaction_level":"niveldesatisfaccion",
                         "last_evaluation":"ultimaevaluacion",
                         "number_project":"numerosdeproyectos",
                         "average_montly_hours":"horasmensualespromedio",
                         "time_spend_company":"tiempoenlaempresa",
                         "Work_accident":"accidentedetrabajo",
                         "left":"abandono",
                         "promotion_last_5years":"promocionultimos5años",
                         "sales":"ventas",
                         "salary":"sueldo"}, inplace= True)
```

```
analisis.info()
```

 ## **<font color="#07a8ed">Visualización de los Datos</font>**

 ### **<font color="#07a8ed">Con Plotly</font>**

```
px.scatter(analisis,
           x="niveldesatisfaccion",
           y="ultimaevaluacion",
           template="gridon",
           color="abandono",
           title="Nivel de Satisfacción vs. Ultima Evaluacion")
```

```
px.scatter(analisis,
           x="niveldesatisfaccion",
           y="ultimaevaluacion",
           color_continuous_scale="mint",
           template="gridon",
           color="abandono",
           labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                   "ultimaevaluacion":"Ultima Evaluación",
                   "abandono":"Abandono"})
```

```
px.histogram(analisis,
             barmode="relative",
             title="Nivel de Satisfacción",
             template="gridon",
             x="niveldesatisfaccion",
             labels={"niveldesatisfaccion":"Nivel de Satisfacción"}).update_layout(bargap=0.2)
```

```
fig = px.histogram(analisis,
                   barmode="relative",
                   title="Nivel de Satisfacción",
                   template="gridon",
                   x="niveldesatisfaccion",
                   labels={"niveldesatisfaccion":"Nivel de Satisfacción"})

fig.update_layout(bargap=0.2)
fig.show()
```

```
fig = px.histogram(analisis,
                   barmode="relative",
                   title="Ultima Evaluación",
                   template="gridon",
                   x="ultimaevaluacion",
                   labels={"ultimaevaluacion":"Ultima Evaluación"})

fig.update_layout(bargap=0.2)
fig.show()
```

```
px.histogram(analisis, x="sueldo",
             title="Sueldos",
             color="sueldo",
             template="gridon")
```

```
px.histogram(analisis, x="sueldo",
             title="Sueldos",
             color="abandono",
             template="gridon")
```

```
px.scatter(analisis,
           x="niveldesatisfaccion",
           y="ultimaevaluacion",
           color="abandono",
           template="gridon",
           marginal_x="histogram",
           marginal_y="histogram",
           labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                   "ultimaevaluacion":"Ultima Evaluación"})
```

```
px.scatter(analisis,
           x="niveldesatisfaccion",
           y="ultimaevaluacion",
           color="abandono",
           template="gridon",
           marginal_x="violin",
           marginal_y="violin",
           labels={"niveldesatisfaccion":'Nivel de Satisfacción',
                   "ultimaevaluacion":"Ultima Evaluación"})
```

```
analisiscorrelacion = analisis.corr(numeric_only=True).round(2)
```

```
analisiscorrelacion
```

```
type(analisiscorrelacion)
```

 ### **<font color="#07a8ed">Con Seaborn</font>**

```
plt.figure(figsize=(9,7))
sns.heatmap(analisiscorrelacion,
            annot=True)

plt.show()
```

```
sns.jointplot(x="niveldesatisfaccion",
              y="ultimaevaluacion",
              data=analisis,
              hue="abandono",
              height=9)

plt.show()
```

```
sns.lmplot(x="niveldesatisfaccion",
           y="ultimaevaluacion",
           data=analisis,
           fit_reg=False,
           hue="abandono",
           height=7)

plt.show()
```

```
sns.displot(analisis.niveldesatisfaccion,
            kde = True,
            height=8)

plt.xlabel("Nivel de satisfacción")
plt.show()
```

```
analisis2 = analisis.copy(deep=False)
```

<p align="justify">
Importantísimo utilizar el parámetro <code>deep = False</code>, ya que si no se hace, toda modificación que se haga sobre el objeto <code>analisis2</code>, tambien se hara sobre su original que en este caso es analisis.

```
analisis2.drop(columns=(["niveldesatisfaccion","accidentedetrabajo",
                        "abandono","promocionultimos5años","ventas","sueldo"]),
               axis=1, inplace=True)
```

```
analisis2.info()
```

```
sns.pairplot(analisis2)
plt.show()
```

```
analisis3 = analisis.copy(deep=False)
```
