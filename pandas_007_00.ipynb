```python
# Crear una Serie de Tiempo con un rango de fechas

fechas = pd.date_range(start="2024-01-01", periods=10, freq="D")
datos = [10, 12, 15, 14, 16, 20, 22, 24, 26, 30]
serie_tiempo = pd.Series(data=datos, index=fechas)
```

 <p align="justify">  El objeto <b>timedelta</b> del m贸dulo <code>datetime</code> representa una diferencia o duraci贸n entre dos fechas u horas. Es especialmente 煤til en aplicaciones de negocios para calcular plazos, vencimientos o intervalos entre eventos.
<br><br>
Una empresa necesita calcular la fecha de vencimiento de facturas emitidas, considerando un plazo de pago est谩ndar de 30 d铆as. Adem谩s, se determina si la fecha de vencimiento cae en fin de semana para ajustar al siguiente d铆a h谩bil.
</p>


```python
from datetime import datetime, timedelta

# Fecha de emisi贸n de una factura

fecha_emision = datetime(2024, 11, 15)
print("Fecha de emisi贸n:", fecha_emision)
```

```python
# Plazo de pago est谩ndar (30 d铆as)

plazo_pago = timedelta(days=30)
plazo_pago
```

```python
# Calcular la fecha de vencimiento

fecha_vencimiento = fecha_emision + plazo_pago
print("Fecha de vencimiento inicial:", fecha_vencimiento)
```

```python
fecha_emision.weekday()  ### cuenta desde 0 el lunes
```

```python
fecha_vencimiento.weekday()
```

```python
# Assuming you want to access the 'day' attribute of fecha_vencimiento:
fecha_vencimiento.day

# If you wanted to change the 'day' attribute of fecha_vencimiento, you could do this:
# (But this creates a new datetime object, it doesn't modify the original)
from datetime import timedelta
new_fecha_vencimiento = fecha_vencimiento + timedelta(days=5 - fecha_vencimiento.day) #Example: Change the day to the 5th
```

```python
new_fecha_vencimiento
```

```python
# Ajustar si el vencimiento cae en fin de semana

if fecha_vencimiento.weekday() >= 5:  # 5 = S谩bado, 6 = Domingo
    fecha_vencimiento += timedelta(days=(7 - fecha_vencimiento.weekday()))
    print("Fecha de vencimiento ajustada al siguiente d铆a h谩bil:", fecha_vencimiento)

```

<p align="justify">
Si la factura se emite el 15 de noviembre de 2024, la fecha inicial de vencimiento ser铆a el 15 de diciembre de 2024. Si esta cae en un fin de semana, se ajustar谩 autom谩ticamente al lunes siguiente.
<br><br>
El c贸digo tambi茅n indica cu谩ntos d铆as faltan hasta el vencimiento.








```python
# Calcular d铆as restantes para el vencimiento

hoy = datetime.now()
dias_restantes = (fecha_vencimiento - hoy).days
print(f"D铆as restantes para el vencimiento: {dias_restantes} d铆as")
```

<p align="justify"> 
El objeto <b>timedelta</b> tambi茅n es 煤til en c谩lculos de fechas complejos, como la gesti贸n de contratos en negocios. A continuaci贸n, se presenta un ejemplo m谩s elaborado:
<br><br>
Una empresa gestiona contratos de servicio con clientes. Cada contrato tiene:
<br>
<ul>
    <li>Una fecha de inicio.</li>
    <li>Una duraci贸n definida (en meses).</li>
    <li>Un per铆odo de gracia de 7 d铆as posterior a la fecha de vencimiento.</li>
</ul>
<br>
<p aling="justify">
El objetivo es calcular:
<ul>
    <li>La fecha de vencimiento del contrato.</li>
    <li>La fecha l铆mite incluyendo el per铆odo de gracia.</li>
    <li>Si la fecha l铆mite ya ha pasado o cu谩nto tiempo queda para su vencimiento.</li>
</ul>
</p>


```python
# Fecha de inicio del contrato

fecha_inicio = datetime(2024, 1, 15)
print("Fecha de inicio del contrato:", fecha_inicio)
```

```python
# Duraci贸n del contrato en meses

duracion_meses = 12
```

```python
# Calcular la fecha de vencimiento (agregando meses)
# Suponiendo que los meses tienen en promedio 30 d铆as

fecha_vencimiento = fecha_inicio + timedelta(days=365)
print("Fecha de vencimiento del contrato:", fecha_vencimiento)
```

```python
# Calcular la fecha de vencimiento (agregando meses)
# Suponiendo que los meses tienen en promedio 30 d铆as

fecha_vencimiento = fecha_inicio + timedelta(days=duracion_meses * 30)
print("Fecha de vencimiento del contrato:", fecha_vencimiento)
```

```python
# Per铆odo de gracia (7 d铆as)

periodo_gracia = timedelta(days=7)
fecha_limite = fecha_vencimiento + periodo_gracia
print("Fecha l铆mite incluyendo per铆odo de gracia:", fecha_limite)
```

```python
from datetime import datetime, timedelta

# ... (rest of your code)

# Determinar estado actual del contrato

hoy = datetime.now() # Use datetime.now() instead of datetime.now()
if hoy > fecha_limite:
    print("El contrato ha vencido. Fecha l铆mite pasada.")
else:
    dias_restantes = (fecha_limite - hoy).days
    print(f"El contrato est谩 vigente. D铆as restantes para la fecha l铆mite: {dias_restantes}")
```

```python
# Determinar estado actual del contrato
from datetime import datetime, timedelta
hoy = datetime.now()
if hoy > fecha_limite:
    print("El contrato ha vencido. Fecha l铆mite pasada.")
else:
    dias_restantes = (fecha_limite - hoy).days
    print(f"El contrato est谩 vigente. D铆as restantes para la fecha l铆mite: {dias_restantes}")
```

```python
# Calcular penalizaci贸n por vencimiento (si aplica)

dias_vencidos = (hoy - fecha_limite).days if hoy > fecha_limite else 0
if dias_vencidos > 0:
    penalizacion_por_dia = 50  # Penalizaci贸n en d贸lares por d铆a de retraso
    penalizacion_total = dias_vencidos * penalizacion_por_dia
    print(f"Penalizaci贸n por {dias_vencidos} d铆as de retraso: ${penalizacion_total:.2f}")
```
